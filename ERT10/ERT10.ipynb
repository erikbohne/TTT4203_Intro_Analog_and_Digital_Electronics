### Task 1
Plot Voltage as a function of current with different values of R

```python
# Import libaries
import numpy as np
import matplotlib.pyplot as plt
```

After doing some measurements with a 9.0V battery (actually 9.45)

```python
R = [100, 300, 1000, 3300, 10000, 33000] # In Ohm
U = [8.69, 9.18, 9.35, 9.41, 9.45,  9.45] # In V
I = [90.2, 28.4, 9.60, 2.89, 0.96, 0.29] # In mA


# Plot the results with regression line
ax1 = plt.subplot(111)

# Plot voltage vs resistance on the first subplot
ax1.plot(I, U, 'o-')
ax1.set_ylabel('Voltage (V)')
ax1.set_xlabel('Current (mA)')
ax1.set_title('Voltage as a function of Current')

# Show the plot
plt.show()
```

### Task 2
**Why does the lower resistances get hotter than the higher ones?**

The lower resistances get hotter than the higher ones because the lower resistances have a higher current going through them. The higher the current, the more energy is being dissipated as heat.

**Try to fit a straight line in the form v = ai + b to the measurement points. What are the values of a and b? What dimensions (units) do these constants have?**

Lets try to fit a straight line to the data points. We can use the `scipy.optimize.curve_fit` function to do this. We will use the `curve_fit` function to fit a line of the form `v = ai + b` to the data points. The `curve_fit` function returns two values, the first is an array of the coefficients of the fit, and the second is the covariance matrix of the fit. We will only use the first value.

```python
# Calculate the slope and intercept
slope, intercept = np.polyfit(I, U, 1)

# Print the slope and intercept
print('Slope (a): ', slope)
print('Intercept (b): ', intercept)
```
